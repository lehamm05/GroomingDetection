## Load dataset 

```python
# import important libraries
import pandas as pd
```

```python
# load dataset
dataset = pd.read_csv('data/dataset.csv')
dataset
```

## Get English

```python
# handle missing values by replacing with empty string
dataset['text'] = dataset['text'].fillna('')  
dataset['text'] = dataset['text'].astype(str)
```

```python
# group by conversation_id
grouped_df = dataset.groupby('conversation_id')['text'].apply(lambda x: ' '.join(x)).reset_index()
```

```python
grouped_df
```

```python
# import langdetect to select all conversation_ids of english conversations
from langdetect import detect
english_conversation = []
for index, row in grouped_df.iterrows():
    try: 
        if detect(row['text'])=='en':
            english_conversation.append(row['conversation_id'])
    except: 
        pass
english_conversation
```

## Load dataset with sentiments and label it

```python
# load dataset
sentiments = pd.read_csv('data/dataset_w_sentiments.csv')
sentiments
```

```python
# load dataset
groomers= pd.read_csv('data/groomer_conversations.csv')
groomers_list = groomers['conversation_id'].to_list()
```

```python
# label grooming conversations based on author IDs
sentiments['label'] = sentiments['conversation_id'].apply(lambda cid: 1 if cid in groomers_list else 0)
```

```python
sentiments
```

```python
print('Grooming_conversations', sentiments[sentiments['label']==1].shape[0])
```

```python
print('Total conversations:', sentiments['conversation_id'].nunique())
pred_conv = sentiments[sentiments['label']==1]
print('Predatory_conversations:', pred_conv['conversation_id'].nunique())
```

```python
english_sentiments = sentiments[sentiments['conversation_id'].isin(english_conversation) | (sentiments['label'] == 1)]
english_sentiments
```

```python
print('Total english conversations:', english_sentiments['conversation_id'].nunique())
pred_conv_eng = english_sentiments[english_sentiments['label']==1]
print('Predatory_conversations:', pred_conv_eng['conversation_id'].nunique())
```

```python
english_sentiments.to_csv('data/dataset_eng.csv', index=False)
```
